```python
from google.colab import drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
drive.mount('/content/drive')
```

```python
data_path = '/content/drive/MyDrive/Colab Notebooks/MiningProcess_Flotation_Plant_Database.csv'
df = pd.read_csv(data_path)
```

```python
df.columns[1:]
```

```python
df['date'] = pd.to_datetime(df['date'])
for column in df.columns[1:]:
  df[column] = df[column].astype(str).str.replace(',', '.', regex=False).astype(float)
```

```python
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix,  annot=True, fmt=".2f", vmin=-1, vmax=1, center=0)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```
